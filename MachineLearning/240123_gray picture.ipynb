```python

color_img = Image.open('fruits.jpg')
color_pixel = np.array(color_img)
plt.imshow(color_pixel)
plt.show()
print(color_pixel.shape)
```

This picture has 426 rows, 640 column and 3 channels like R, G, B.
※.jpg has 3 channels
  .png has 4 channels.

```python
gray_pixel = color_pixel
#gray_pixel = color_pixel.copy()
for y in range(gray_pixel.shape[0]): #row
    for x in range(gray_pixel.shape[1]):#column
        gray_pixel[y, x] = np.mean(gray_pixel[y, x])

plt.imshow(gray_pixel)
plt.show()
plt.imshow(color_pixel)
plt.show()
```

it has problem. 'color_pixel' shows picture with gray color.
because 'color_pixel' reference with 'gray_pixel'. 
'gray_pixel = color_pixel' means shallow copy.
So, you have to deep copy and change the word like 'gray_pixel = color_pixel.copy()'

```python
#흑백 이미지를 2차원표현
gray_2D_pixel = gray_pixel[:, :, 0] # x값 전체, y값 전체, RGB 중 R 값만 가져옴 
print(gray_2D_pixel.shape)
plt.imshow(gray_2D_pixel)
plt.show()
```

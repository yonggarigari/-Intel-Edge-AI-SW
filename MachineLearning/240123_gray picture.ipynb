```python
gray_pixel = color_pixel
#gray_pixel = color_pixel.copy()
for y in range(gray_pixel.shape[0]): #row
    for x in range(gray_pixel.shape[1]):#column
        gray_pixel[y, x] = np.mean(gray_pixel[y, x])

plt.imshow(gray_pixel)
plt.show()
plt.imshow(color_pixel)
plt.show()
```
